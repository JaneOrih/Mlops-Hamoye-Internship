This notebook contains the lab/project work for Hamoye Internship (stage_B)

Dataset:Energy Prediction data

The data set is at 10 min for about 4.5 months. The house temperature and humidity conditions were monitored with a ZigBee wireless sensor network. Each wireless node transmitted the temperature and humidity conditions around 3.3 min. Then, the wireless data was averaged for 10 minutes periods. The energy data was logged every 10 minutes with m-bus energy meters. Weather from the nearest airport weather station (Chievres Airport, Belgium) was downloaded from a public data set from Reliable Prognosis (rp5.ru), and merged together with the experimental data sets using the date and time column

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# to display all columns
pd.pandas.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('Downloads/energydata_complete (1).csv')
df.head()
```

```python
#12


x= df['T2'].values.reshape(-1,1)
y =df['T6'].values.reshape(-1,1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=42)

#linear regression training with T2 and T6
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
y_pred=lin_reg.predict(x_test)

#r_square metric
from sklearn.metrics import r2_score
r2=r2_score(y_test, y_pred)
print(f'r_square: {round(r2,2)}')
```

```python
#13, 14, 15, 16

df=df.drop(columns=['date', 'lights'])

# Normalizing the dataset
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
norm_df=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)

x=norm_df.drop(columns=['Appliances'])
y=norm_df['Appliances']

#dividing dataset into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=42)

# training using linear Regression
from sklearn.linear_model import LinearRegression
lin_regr=LinearRegression()
lin_regr.fit(x_train,y_train)

y_pred=lin_regr.predict(x_test)

# Question 13
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
print(f'Mean Absolute Error : {round(mae,2)}')

# Question 14
# RSS
rss=np.sum(np.square(y_test-y_pred))
print(f'Residual Sum of Squares : {round(rss,2)}')

# Question 15
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print(f'Root Mean Squared Error : {round(rmse,3)}')

# Question 16
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print(f'Coefficient of Determination : {round(r2,2)}')
```

```python
def get_wt_df(model,feat,col_name):
    wt=pd.Series(model.coef_,feat.columns).sort_values()
    wt_df=pd.DataFrame(wt).reset_index()
    wt_df.columns=['Features',col_name]

    return wt_df
```

```python
# Question 17
get_wt_df(lin_regr,x_train,'Linear model').sort_values(by=['Linear model'],ascending=True)
```

therefore; RH_1	and RH_2 are the features with maximum and minimum weight

```python
# Question 18

# Ridge reegression
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=0.4)
ridge.fit(x_train,y_train)

y_pred=ridge.predict(x_test)

from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print(f'Root Mean Squared Error : {round(rmse,3)}')
```

there is no change in the Root Mean Squared Error for ridge and linear model

```python
# Question 19

# Lasso Regression
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.001)
lasso.fit(x_train,y_train)

get_wt_df(lasso,x_train,'Lasso_wt').sort_values(by=['Lasso_wt'],ascending=True)
```

4 features have non zero weights

```python
# Question 20

y_pred_lasso=lasso.predict(x_test)
rmse_lasso=np.sqrt(mean_squared_error(y_test,y_pred_lasso))
print(f'Root Mean Squared Error : {round(rmse_lasso,3)}')
```
